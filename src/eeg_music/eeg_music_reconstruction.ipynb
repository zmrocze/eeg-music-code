```python
from pathlib import Path
from eeg_music.data import EEGMusicDataset
ds = EEGMusicDataset.load_ondisk(Path("./datasets/bcmi_preprocessed/bcmi_pre_60ch/"))
len(ds)
```

```python
from eeg_music.data import MappedDataset, MelParams, prepare_trial


mds = MappedDataset(ds, lambda t: 
  prepare_trial(t, eeg_resample=None, 
    wav_resample=32768, # 64 * 512,
    apply_mel=MelParams(
      n_mels=64, 
      fmax=10240.0))) # 8192.0
mds[0].music_data.mel.shape
```

```python
from fractions import Fraction
from eeg_music.data import StratifiedSamplingDataset


smds = StratifiedSamplingDataset(mds, 10, Fraction(1,1))
smds[0].music_data.mel.shape
```
