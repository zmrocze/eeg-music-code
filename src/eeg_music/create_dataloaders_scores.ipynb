```python
from pathlib import Path

from eeg_music.bcmi import BCMICalibrationLoader, BCMITrainingLoader, BCMIScoresLoader
from eeg_music.data import EEGMusicDataset
from eeg_music.data import copy_from_dataloader_into_dir
import tempfile
import matplotlib.pyplot as plt
from eeg_music.data import MappedDataset, MelParams, prepare_trial, rereference_trial
from fractions import Fraction
from eeg_music.data import StratifiedSamplingDataset

# # %load_ext autoreload
# %autoreload 2
```

```python
def create_combined_training_scores_dataset(
  bcmi_root_path: str, new_dataset_save_dir: Path,
) -> EEGMusicDataset:
  """Create an EEGMusicDataset that combines BCMI training and calibration data."""

  training_loader = BCMIScoresLoader(f"{bcmi_root_path}/bcmi-scores", dataset_name="bcmi-scores")
  training_loader.load_all_subjects()
  copy_from_dataloader_into_dir(training_loader, new_dataset_save_dir)
  del training_loader
  combined_dataset = EEGMusicDataset.load_ondisk(new_dataset_save_dir)

  return combined_dataset

def create_training_dataset(
  bcmi_root_path: str, new_dataset_save_dir: Path,
) -> EEGMusicDataset:
  """Create an EEGMusicDataset that combines BCMI training and calibration data."""

  training_loader = BCMITrainingLoader(f"{bcmi_root_path}/bcmi-training", dataset_name="bcmi-training")
  training_loader.load_all_subjects()
  copy_from_dataloader_into_dir(training_loader, new_dataset_save_dir)
  del training_loader
  combined_dataset = EEGMusicDataset.load_ondisk(new_dataset_save_dir)

  return combined_dataset
```

```python
temp_save_dir = Path(tempfile.mkdtemp(prefix="bcmi_combined_dataset_"))
combined_dataset = create_training_dataset("./datasets/bcmi", temp_save_dir)
```

```python
from eeg_music.onset_conversion import trial_wavraw_to_noteonsets
# from eeg_music.data import prepare_trial

filtered_ds = EEGMusicDataset()
filtered_ds.df = combined_dataset.df.sample(n=150, random_state=420).reset_index(drop=True)
filtered_ds.music_collection = combined_dataset.music_collection

one_sec_ds = MappedDataset(
  filtered_ds,
  lambda t: trial_wavraw_to_noteonsets(
    prepare_trial(
      t,
      eeg_resample=512,
      max_len=1.0,
    )  # type: ignore
  )
)

len(one_sec_ds)
```

```python
from eeg_music.eegnet import has_onset_in_window

# Configuration for the onset detection window
# window_start = 3*32  # adjust as needed
# window_end = 5*32  # adjust as needed (e.g., 1 second at 256 Hz)
# eeg_sample_rate = 256  # adjust to your dataset's sample rate
window_start = 3*64  # adjust as needed
window_end = 5*64  # adjust as needed (e.g., 1 second at 256 Hz)
eeg_sample_rate = 512  # adjust to your dataset's sample rate

# Count positive and negative trials
positive_count = sum(
    has_onset_in_window(trial.music_data.get_music(), window_start, window_end, eeg_sample_rate) # type: ignore
    for trial in (one_sec_ds[i] for i in range(len(one_sec_ds)))
)
negative_count = len(one_sec_ds) - positive_count

print(f"Positive trials: {positive_count}")
print(f"Negative trials: {negative_count}")
print(f"Total trials: {len(one_sec_ds)}")
print(f"Positive ratio: {positive_count / len(one_sec_ds):.2%}")
```
